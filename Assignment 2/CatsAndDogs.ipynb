```
#data_files = glob.glob("/content/drive/My Drive/Deep Learning/Assignment/Assignment 2/CatDog/Cat/*.jpg")
data_files = glob.glob("CatDog/PetImages/Cat/*.jpg")
```

```
data_files
```

```
from PIL import Image 
import numpy as np
```

```
catimages=[]
for files in data_files:
  img=Image.open(files)
  img=img.resize((28,28))
  img=img.convert(mode='L')
  catimages.append([img,0])

  
```

```
#dog_files = glob.glob("/content/drive/My Drive/Deep Learning/Assignment/Assignment 2/CatDog/Dog/*.jpg")
dog_files = glob.glob("CatDog/PetImages/Dog/*.jpg")
```

```
dog_files
```

```
for files in dog_files:
  img=Image.open(files)
  img=img.resize((28,28))
  img=img.convert(mode='L')
  catimages.append([img,1])
```

```
import random
random.shuffle(catimages)
```

```
x=[]
y=[]
```

```
for image,label in catimages:
  x.append(np.array(image))
  y.append(label)
```

```
x=np.array(x)
```

```
x=x/255
```

```
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.33, random_state=42)
```

```
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import Dropout
```

```
model=Sequential()
model.add( Flatten(input_shape=(28,28)))
model.add(Dense(128,activation='relu'))
model.add(Dense(70,activation='relu'))
model.add(Dense(2,activation='softmax',))
```

```
model.summary()
```

```
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```
model.fit(xtrain, ytrain, epochs=60)
```

```
loss, accuracy=model.evaluate(xtest,ytest)
```
