```
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```
dataset=pd.read_csv('/content/gdrive/My Drive/Deep Learning/Assignment/Assignment 2/sonar.csv',header=None)
```

```
# Load the values
values = dataset.values
```

```
# Separate the x and y values
xvalues = values[:,0:60].astype(float)
yvalues = values[:,60]
```

```
# Define Lable encoder
enc = LabelEncoder()
enc.fit(yvalues)
enc_y = enc.transform(yvalues)
```

```
# Define the base model
def create_baseline():
  model = Sequential()
  model.add(Dense(60, input_dim=60, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))
  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
  return model
	
```

```
# Define the estimator
estimator = []
estimator.append(('standardize', StandardScaler()))
estimator.append(('mlp', KerasClassifier(build_fn=create_baseline, epochs=100, batch_size=5, verbose=0)))
```

```
# Setup a pipeline
pip = Pipeline(estimator)
```

```
# Execute the model using stratified k fold
kfold = StratifiedKFold(n_splits=10, shuffle=True)
```

```
results = cross_val_score(pip, xvalues, enc_y, cv=kfold)
```

```
print("Standardized: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```
