```

trainset, validationdata, testset = tfds.load(
    name="imdb_reviews", 
    split=('train[:60%]', 'train[60%:]', 'test'),
    as_supervised=True)
```

```
trainbatch, trainlabelsbatch = next(iter(trainset.batch(10)))
trainbatch
```

```
trainlabelsbatch
```

```
embedding = "https://tfhub.dev/google/tf2-preview/gnews-swivel-20dim/1"
hublayer = hub.KerasLayer(embedding, input_shape=[], 
                           dtype=tf.string, trainable=True)
hublayer(trainbatch[:3])
```

```
model = tf.keras.Sequential()
model.add(hublayer)
model.add(tf.keras.layers.Dense(16, activation='relu'))
model.add(tf.keras.layers.Dense(1))

model.summary()
```

```
model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```
history = model.fit(trainset.shuffle(10000).batch(512),
                    epochs=20,
                    validation_data=validationdata.batch(512),
                    verbose=1)
```

```
results = model.evaluate(testset.batch(512), verbose=2)

for name, value in zip(model.metrics_names, results):
  print("%s: %.3f" % (name, value))
```
